```python
# 실습문제 1번

import csv
import matplotlib.pyplot as plt

f = open('seoul1.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
high = []
low = []
for row in data:
    if row[-1] != '':
        if(row[0].split('-')[1] == '03') & (row[0].split('-')[2] == '10'):
            high.append(float(row[-1]))
            low.append(float(row[-2]))
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
plt.title('생일날 서울 지역 최고기온과 최저기온')
plt.plot(high, color='red', label='최고 기온')
plt.plot(low, color='blue', label='최저 기온')
plt.legend()
plt.show()
plt.close()
```

```python
# 실습문제 2번

import csv
import matplotlib.pyplot as plt

f = open('seoul1.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
janu = []
augu = []
for row in data:
    if row[0].split('-')[1] == '01':
        janu.append(float(row[-1]))
    if row[0].split('-')[1] == '08':
        augu.append(float(row[-1]))
plt.rc('font', family='Malgun Gothic')
plt.title('서울 지역 1월과 8월의 최고 기온 히스토그램')
plt.hist(janu, bins=100, color='blue', label='1월')
plt.hist(augu, bins=100, color='red', label='8월')
plt.legend()
plt.show()
plt.close()
```
