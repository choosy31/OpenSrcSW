```python
# 실습문제 1번
import csv

f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f,delimiter=',')
header = next(data)
max = -999
max_day = ''
min = 999
min_day = ''
range = 0
range_day = ''
for row in data:
    row[-1] = float(row[-1])
    row[-2] = float(row[-2])
    diff = row[-1] - row[-2]
    if diff > range:
        range = diff
        range_day = row[0]
    if float(row[-2]) < min:
        min = row[-2]
        min_day = row[0]
    if float(row[-1]) > max:
        max = row[-1]
        max_day = row[0]
print("일교차가 가장 큰 날:")
print(range_day)
print("최저기온이 가장 낮은 날")
print(min_day)
print("최고기온이 가장 좋은 날")
print(max_day)
f.close()
```

```python
# 실습문제 2번
import csv

f = open('subwayfee.csv', 'r', encoding='cp949')
data = csv.reader(f,delimiter=',')
header = next(data)
turn = []
b= ''
for row in data:
    row[-1]=row[-1].replace(",","")
    row[-2]=row[-2].replace(",","")
    a = int(row[-1])+int(row[-2])
    row.append(a)
    if (turn != list()) & (row[1] != b):
        for rw in turn:
            for i in range(turn.index(rw)+1, len(turn)):
                if(turn[i][-1] > turn[turn.index(rw)][-1]):
                    for n in range(7):
                        turn[turn.index(rw)][n],turn[i][n] = turn[i][n],turn[turn.index(rw)][n] 
        print("----",turn[0][1],"----")
        print("이용객이 많은 역 순위:")
        print("1.",turn[0][3],"(",turn[0][-1],"명)")
        print("2.",turn[1][3],"(",turn[1][-1],"명)")
        print("3.",turn[2][3],"(",turn[2][-1],"명)")
        print("이용객이 적은 역 순위:")
        print("1.",turn[-1][3],"(",turn[-1][-1],"명)")
        print("2.",turn[-2][3],"(",turn[-2][-1],"명)")
        print("3.",turn[-3][3],"(",turn[-3][-1],"명)")
        print()
        turn = []
    b = row[1]
    turn.append(row)
f.close()
```
